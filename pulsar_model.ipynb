### 1). INTEGRATED PROFILE

pulsars produce periodic pulsation signals that are too weak to detect.Integrated pulsar signals with respect to its rotational period.This increase signal quality and form integrated pulsar profilepulsars are distinguished by integrated pulse profile as all pulsar's integrated profile have unique shape.So,it works as pulsar's fingerprint

Over the whole electromagnetic spectrum from radio to the very high energy gamma-rays detected by air shower telescopes, pulsars radiate beams giving pulse profiles which are unique signatures differing from pulsar to pulsar. For some individual pulsars,very similar profiles are observed over the whole spectrum. The majority of pulsars are observable only in radio; here the individual pulses are often very variable and it is by averaging the pulse shapes of many, sometimes hundreds, of pulses that the characteristic shape is seen.


### 2). DM-SNR CURVE

 Radio waves emitted from pulsars reach earth after travelling long distances in space which is filled with free electrons. Since radio waves are electromagnetic in nature, they interact with these electrons, this interaction results in slowing down of the wave. The important point is that pulsars emit a wide range of frequencies, and the amount by which the electrons slow down the wave depends on the frequency. Waves with higher frequency are sowed down less as compared to waves higher frequency. i.e. lower frequencies reach the telescope later than higher frequencies. This is called dispersion.

## Training Algorithm

```python
sns.heatmap(data.corr(),annot=True)
```

for target_class variables with high correlation are *Excess kurtosis with the integrated profile* **0.78**, *Skewness of the integraged profile* **0.71** *Standard deviation of the DM-SNR curve* **0.47** and *Mean of the DM-SNR curve* **0.4**

```python
## selecting important features and target variable

features= data[[" Excess kurtosis of the integrated profile"," Skewness of the integrated profile"," Standard deviation of the DM-SNR curve"," Mean of the DM-SNR curve"," Skewness of the DM-SNR curve"]]
target=data[["target_class"]]
```

```python
## splitting data into training and testing 

from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(features,target,train_size=0.7,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

log_reg=LogisticRegression()
log_reg.fit(xtrain,ytrain)
ypred_log_reg=log_reg.predict(xtest)
sns.heatmap(metrics.confusion_matrix(ytest,ypred_log_reg),annot=True)
plt.xlabel("predicted")
plt.ylabel("actual")
print("Accuracy= "+str(metrics.accuracy_score(ytest,ypred_log_reg)))
```

#### accuracy on test data=0.9789

```python
x=log_reg.predict_proba(xtrain)[::,1]
fpr,tpr,_=metrics.roc_curve(ytrain,x)
auc=metrics.roc_auc_score(ytrain,x)
plt.plot(fpr,tpr,label="auc= "+str(auc))
plt.legend()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=0,n_jobs=-1)
mod_rf=rf.fit(xtrain,ytrain)
ypred_rf=mod_rf.predict(xtest)
```

```python
print("CONFUSION MATRIX: \n"+str(metrics.confusion_matrix(ytest,ypred_rf) ))
print("\nACCURACY: "+str(metrics.accuracy_score(ytest,ypred_rf)) )
```

There is no big difference in accuracy of **Random Forest(0.97765)** and **Logistic Regression(0.97893)**

You can see that amount of False-Positive and False-Negative is lesser than logistic regression model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

std=StandardScaler()
std.fit(xtrain)

xtrain_std=std.transform(xtrain)
xtest_std=std.transform(xtest)

knn=KNeighborsClassifier(n_neighbors=5,n_jobs=-1)
knn.fit(xtrain_std,ytrain)
ypred_knn=knn.predict(xtest_std)
```

```python
print("CONFUSION MATRIX: \n"+str(metrics.confusion_matrix(ytest,ypred_knn)))
print("\nACCURACY: "+str(metrics.accuracy_score(ytest,ypred_knn)))
```

here accuracy=0.97924 which is the highest in all the three models
